```python
import pandas as pd

def prepare_data():
    df_main = pd.read_csv(r'C:\work\DrugDiscovery\main_git\XAI_Chem\data\init_data\pKa_Prediction_Starting data_2024.01.25.csv', index_col=0)
    # df_main = pd.read_csv(r'data\init_data\pKa_Prediction_Starting data_2024.05.07.csv', index_col=0)

    SMILES_to_fgroup = {}
    SMILES_to_identificator = {}
    SMILES_to_cycle_type = {}
    for index, row in df_main.iterrows():
        SMILES = row['Amides for LogP']
        if pd.isnull(SMILES):
            continue
        
        SMILES_to_fgroup[SMILES] = row['F group']
        SMILES_to_identificator[SMILES] = row['identificator']
        SMILES_to_cycle_type[SMILES] = row['Framework']
    
    return SMILES_to_fgroup, SMILES_to_identificator, SMILES_to_cycle_type
```

```python
SMILES_to_fgroup, SMILES_to_identificator, SMILES_to_cycle_type = prepare_data()

df_train = pd.read_csv(r'C:\work\DrugDiscovery\main_git\XAI_Chem\data\logP_lipophilicity_data\gnn_cv\train.csv', index_col=0)
df_test = pd.read_csv(r'C:\work\DrugDiscovery\main_git\XAI_Chem\data\logP_lipophilicity_data\gnn_cv\test.csv', index_col=0)
df = pd.concat([df_train, df_test], axis=0)

df_for_train = pd.read_csv(r'C:\work\DrugDiscovery\main_git\XAI_Chem\data\logP_lipophilicity_data\gnn_cv\train_test_for_SMILES_per_model.csv')
```

```python
from rdkit import Chem

for index_train, row_train in df.iterrows():
    SMILES = row_train['Smiles']
    canon_SMILES = Chem.CanonSmiles(SMILES)

    is_exist_in_train = False
    smiles_non_f = None
    for index, row in df_for_train.iterrows():
        if row['Smiles'] == canon_SMILES:
            is_exist_in_train = True
            smiles_non_f = row['non-F Smiles']
            break
    
    model_name = f"{index}_{canon_SMILES}_{smiles_non_f}.pkl"
    print(model_name, index_train, index)
```
